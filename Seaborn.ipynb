### 1: Distribion Plots: Histogram, KDE (Kernel Density Estimate), Displot

```python
tips = sns.load_dataset('tips') #load dataset tips from the library 
tips.head()
```

```python
sns.histplot(data = tips['total_bill'], color = 'blue');
```

```python
#Check sum valuse to assess the order of the columns
tips['total_bill'].value_counts().sort_values(ascending = False)
```

### KDE Plot 
A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analagous to a histogram.
KDE represents the data using a continuous probability density curve in one or more dimensions
( Just looks like a graph in high school to find max min, the special points on the graph )

```python
sns.kdeplot(tips['total_bill']);
```

### Displot
Informative distributional summaries

Statistical analyses require knowledge about the distribution of variables in your dataset.
The seaborn function displot() supports several approaches to visualizing distributions.

These include classic techniques like histograms and computationally-intensive approaches like kernel density estimation



```python
fig = sns.displot(tips['tip'], kde = True, color = 'purple')
fig.set( xlabel = 'The money tips for the staff' ,
       ylabel = 'The count',
       title = 'The graph mix between KDE and Displot')
```


## Box Plot
Box Plot is a good indication of how the values in the data are spread out ("minium", first quartile (Q1), median, third quartile (Q3), and “maximum”)

Identify outliers
Check if data is symmetrical
Determine how tightly data is grouped

```python
sns.boxplot(tips, x = 'day', y = 'total_bill', hue = 'sex', palette = 'Blues' );
plt.legend(loc = 0)
```

## Facet grid 
FacetGrid can be drawn with up to three dimensions: row, col, and hue.

The main approach for visualizing data on this grid is with the FacetGrid.map() method.

```python
tips.head()
```

```python
fig1 = sns.FacetGrid(tips, row = 'smoker', col = 'time', )
fig1.map(sns.scatterplot, 'total_bill', 'tip');
```
