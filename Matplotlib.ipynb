# Object-Oriented API 

```python
# prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. setup plot
fig, ax = plt.subplots(figsize=(5,5)); #Figure size = Width & Height of the plot
ax.plot(x,y);

#3. Plot the data
ax.plot(x,y);
ax.set(title = "Moi quan he x va y",    #Thay vi dung plt.title...
      xlabel = "Don vi x",
      ylabel = "DOn vi y theo x")
```

```python
##00 API
x = np.random.random(size = 100); #Radom tao 1 day x, y gia tri tu 0 den 1 voi 100 gia tri
y = np.random.random( size = 100);
sizes = 100 * np.random.random(size = 100);  #tao ra kich thuoc bang cach nhan 100 * gia tri tu 0 den 1 (donvi: pixel)
colors = np.random.random(size = 100) #color ngay nhien tu 1-100
fig, ax = plt.subplots()  #method for OO
# s = sizes, colors, cmap -> y nghia: (colormaps): anh xa gia tri vao bang mau cu the
img1 = ax.scatter(x, y, s=sizes, c=colors, cmap = 'viridis' , alpha = 0.7)
fig.colorbar(img1) #tao thanh colorbar
```
