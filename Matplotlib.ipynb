## Subplots

```python
#Plot data each axis
fig, ax = plt.subplots( nrows = 2,  #setup kich thuoc cho subplots
                      ncols = 2, 
                      figsize = (10,10) ) #Chia khung thanh ma tran 2x2, thanh 
# 4 bieu do con, figsize: Kich thuoc khng hinh (10,10) inch

#Truy cap bieu do giong voi list ax[hang, cot]. -> Object Oriented
#Chart 1 (Hang 0, cot 0)
ax[0,0].plot(x, x/2, color = 'red', label = 'X')
ax[0,0].legend()
ax[0,0].set(title = 'Do thi y va x theo truc toa do Decartes')

#Chart 2 (Hang 0, cot 1)
color = np.random.random(size =100)
size = np.random.random(size = 100) * 300
ax[0,1].scatter( np.random.random(size =100), np.random.random(size = 100),
                 c = color, cmap = 'ocean')
ax[0,1].set(title = 'Bag phan bo tan suat chung vi khuan E.coli');


#Chart 3(Hang 1, cot 0)
soft_drink = {'coke':8, 'pepsi':10, 'mirinda':9, 'aquafina':5}
ax[1,0].bar(soft_drink.keys(), soft_drink.values(), color = 'skyblue')
ax[1,0].set(title = 'Gia ban mot so loai nuoc ngot',
           xlabel = 'Loai nuoc ngot')

#Chart 4(Hang 1, cot 1)
ax[1,1].hist(student_height, bins = 35, color = 'steelblue')
ax[1,1].set(title = 'Bang phan bo chieu cao cua sinh vien dai hoc',
           xlabel = 'Chieu cao (donvi: cm)')

plt.tight_layout()
```

## Project Data Visualization - California Cities

```python
#Setup library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read CSV file
file = pd.read_csv(r'C:\Users\NA\Documents\Materials\Project\Data Science Materials\Data sample\california_cities.csv')
file.head()
```

```python
#ectract data from csv file to tabular (array)
#Latd: Vi do, Longd: Kinh do

lat, lod = file['latd'], file['longd']
pop, area = file['population_total'], file['area_total_km2']

#scatter plot to display population and size of area
fig, ax = plt.subplots(figsize = (8,6))
img1 = ax.scatter(lat, lod, c = np.log10(pop), cmap ='ocean',
          s=area, linewidths =0)


ax.set( xlabel = 'longlatitude',
      ylabel = 'latitude',
      title = 'California Cities: Population and Area Distribution');

img1.set_clim(3,7) #fig hay ax khong su dung duoc vi chung chi quan ly khong gian ben ngoai, 
#di sau hon can doi tuong co ten cu the (img1) de co the tuy chinh mau sac.
fig.colorbar(img1, label = 'log$_{10}$(population)') # Tao thanh colorbar va dat label cho cbar

#make a notice tabular:
listarea = [50,110, 300, 500]
for area in listarea:
    ax.scatter([], [], 
               s=area,          # Kích thước chấm
               c='steelblue',           # Màu đen (k) để phân biệt với màu dân số
               alpha=0.3,       # Độ trong suốt nhẹ
               label= str(area) + ' km$^2$') #

# 3. Hiển thị bảng chú thích với tùy chỉnh khoảng cách
ax.legend(title='City Area', 
            labelspacing = 1,
            frameon = False
          );

import seaborn as sns
sns.set() # Kích hoạt giao diện Seaborn


#li do khong dung legend() vi chung chi hien thi nhung thu co san tren map, vidu nhu hien thi
#mot cham co dinh (fixed point) khong tien cho viec xem ban do tra cuu
```
