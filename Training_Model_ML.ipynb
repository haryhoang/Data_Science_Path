# Training Machine Learning for project Housing Price Prediction

## Problem Definition
Goal: predict the sale price for each house

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Upload data
data = pd.read_csv("./train.csv", index_col = "Id")
data.head()
```

```python
data.columns
```

## Feature Selections
- Choose Features to train ML model
- Need to use "Feature Engineering"

```python
features = ["LotArea", "YearBuilt", "1stFlrSF", "2ndFlrSF", "FullBath", "BedroomAbvGr", 
           "TotRmsAbvGrd"]
data[features]
```

## Spliting the datasheet
- 'data' :dataset
- 'X' : data[features]
- 'Y' : target variable: Saleprices

```python
X = data[features] # features is a external column, use data[..]
X
```

```python
Y = data["SalePrice"] #SalePrice is an internal column, so use syntax data[".."]
Y
```

```python
#Split the dataset into X train and Y train to find the relations.
from sklearn.model_selection import train_test_split
X_train, X_valid, Y_train, Y_valid = train_test_split(X, Y, train_size =0.8, random_state = 0)
```

```python
print (X_train, X_valid) #X_train = 0.8 with X_valid = 0.2
```

## Training Machine Learning Model

```python
from sklearn.tree import  DecisionTreeRegressor
data_model = DecisionTreeRegressor( random_state = 1)
```

```python
# Fit training data into model
data_model.fit(X_train, Y_train)
```

```python
# Try with the data from X_valid to check the accuracy
Y_predict = data_model.predict(X_valid.head(5))
Y_predict
```

```python
pd.DataFrame({"Y_realitic":Y_valid.head(5), "Y_predict":Y_predict})
```

## Another model for Prediction

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
rf_model = RandomForestRegressor(random_state = 1)
rf_model.fit(X_train, Y_train)
```

```python
Y_predict_with_rf = rf_model.predict(X_valid)
Y_predict_with_rf
```

```python
pd.DataFrame({"Y_train":Y_predict_with_rf, "Y_realitic":Y_valid})
```
