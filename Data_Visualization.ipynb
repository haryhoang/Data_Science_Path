## California Population and area 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#read csv file
data = pd.read_csv(r'C:\Users\NA\Documents\Materials\Project\Data Science Materials\Data sample\California_cities.csv')
data.head()
```

```python
pop, area = data['population_total'], data['area_land_km2']
lat, lod = data['latd'], data['longd'] #latd vido, longd: kinh do
name = data['city']
fig, ax = plt.subplots(nrows = 2, 
                      ncols = 2, 
                      figsize = (8,6))
img1 = ax[0,0].scatter( lat, data.elevation_m, s= area, cmap = 'viridis', c=np.log(area))
ax[0,0].set( xlabel = 'Latitude',
            ylabel = 'Elevation (unit = meters',
             title = 'Distribution the population and area')
fig.colorbar(img1, label = 'Area')
water_area = data['area_water_percent']
ax[0,1].bar( np.log(area), (pop))
```
